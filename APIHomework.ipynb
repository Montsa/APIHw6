```python
import requests
import json
from config import api_key
import time
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import csv
```

```python
#create random list of 550 latitudes and logitutdes
lat = []
lon = []

for i in range(0,1500):
    x = np.random.uniform(0,90)
    lat.append(round(x,3))

for j in range(0,1500):
    y = np.random.uniform(0,180)
    lon.append(round(y,3))

```

```python
#create geocoordinate pairs
df = pd.DataFrame(
    {'lat': lat,
     'lon': lon
    })
df.head()
```

```python
responses = []
print_log = []

for ind in df.index:
    url = "http://api.openweathermap.org/data/2.5/weather?"
    params = {
            "appid": api_key,
            "lat": df['lat'][ind],
            "lon": df['lon'][ind],
            "units": "imperial"
}
    weather_data = requests.get(url, params=params).json()
    id = weather_data.get("id")
    city = weather_data.get("name")
    if id != 0:
        print_log.append(print(f'{ind}, {id}, {city}')) 
        responses.append(weather_data)
    else:
        next
```

```python
print(json.dumps(responses, indent=4, sort_keys=True))
```

```python
#Did not capture scatter plot values in variable during api call. Create DataFrame to access values
#through iteration without needing to do another api call
df_weather_data = pd.DataFrame(responses)

# save data to csv, need to specify file path
df_weather_data.to_csv()
```

```python
#create variable to hold scatter plot variables
lat = []
temp =[]
hum = []
cloud = []
wind = []

#iterate through Dataframe to store scatter plot values
for row in df_weather_data.index:
    lat.append(df_weather_data["coord"][row]["lat"])
    temp.append(df_weather_data["main"][row]["temp"])
    hum.append(df_weather_data["main"][row]["humidity"])
    cloud.append(df_weather_data["clouds"][row]["all"])
    wind.append(df_weather_data["wind"][row]["speed"])
    
```

```python
# create a data frame from lat, temp, hum, cloud, wind
scatter_dict = {
    "lat": lat,
    "temp": temp,
    "hum": hum,
    "cloud": cloud,
    "wind": wind
}
df_scatter = pd.DataFrame(scatter_dict)
df_scatter.head()
```

```python
#Scatter plot Temperature (F) vs. Latitude

plt.scatter(df_scatter["lat"], df_scatter["temp"], marker="o")

# Incorporate the other graph properties
plt.title("Temperature (F) vs. Latitude")
plt.ylabel("Temperature (Farenheit)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("TemperatureF_vs_Latitude.png")

# Show plot
plt.show()
```

```python
#Scatter plot Humidity (%) vs. Latitude

plt.scatter(df_scatter["lat"], df_scatter["hum"], marker="o")

# Incorporate the other graph properties
plt.title("Humidity (%) vs. Latitude")
plt.ylabel("Humidity")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Humidity%_vs_Latitude.png")

# Show plot
plt.show()
```

```python
#Scatter plot Cloudiness (%) vs. Latitude

plt.scatter(df_scatter['lat'],df_scatter["cloud"], marker="o")

# Incorporate the other graph properties
plt.title("Cloudiness (%) vs. Latitude")
plt.ylabel("Cloudiness")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Cloudiness%_vs_Latitude.png")

# Show plot
plt.show()
```

```python
#Scatter plot Wind Speed (mph) vs. Latitude

plt.scatter(df_scatter['lat'],df_scatter["wind"], marker="o")

# Incorporate the other graph properties
plt.title("Wind Speed (mph) vs. Latitude")
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("WindSpeedmph_vs_Latitude.png")

# Show plot
plt.show()
```
